```python
import numpy as np 
import sklearn as sk
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split
X,Y=make_regression(n_samples=100,n_features=1,noise=10)
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
# make a linear regression model from scratch
class Linear_reg():
    def __init__(self):
        self.coef_=None
        self.intercept_=None
    def fit(self,X , Y):
        num=0
        den=0
        for i in range(X.shape[0]):
            num+=(X[i]-np.mean(X))*(Y[i]-np.mean(Y))
            den+=(X[i]-np.mean(X))**2
        self.coef_=num/den
        self.intercept_=np.mean(Y)-self.coef_*np.mean(X)
    def predict(self,X):
        return self.coef_*X+self.intercept_



model=Linear_reg()
model.fit(X_train,Y_train)
model.coef_,model.intercept_
plt.scatter(X_train,Y_train)
plt.plot(X_train,model.predict(X_train),color='red')
model.coef_,model.intercept_
        
```

```python
# sklearn linear regression
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

model=LinearRegression()
model.fit(X_train,Y_train)
print(model.coef_,model.intercept_)
y_pred_lr=model.predict(X_test)

plt.scatter(X,Y,color='blue',label='Data points')
plt.plot(X_test,y_pred_lr ,color='red',label='Regression line')
```

```python
# SGD linear regression
from sklearn.linear_model import SGDRegressor
model=SGDRegressor(max_iter=100,learning_rate='constant',eta0=0.01)
model.fit(X_train,Y_train)
print(model.coef_,model.intercept_)
y_pred_sgd=model.predict(X_test)




plt.scatter(X,Y,color='blue',label='Data points')
plt.plot(X_test,y_pred_lr,color='red',label='LR',linewidth=3)
plt.plot(X_test,y_pred_sgd,color='green',label='SGD')
plt.legend()
plt.grid()
plt.title('Linear Regression vs SGD Regression')
```

```python
# SGD regressor from scratch 
class SGD_reg():
    def __init__(self,learning_rate=0.01,epoch=100):
        self.lr=learning_rate
        self.epoch=epoch
        self.m=1
        self.b=100
    def fit(self,X,y):
        # calcualte the b using GD
        for i in range(self.epoch):
            loss_slope_b = -2 * np.sum(y - self.m*X.ravel() - self.b)
            loss_slope_m = -2 * np.sum((y - self.m*X.ravel() - self.b)*X.ravel())
            
            self.b = self.b - (self.lr * loss_slope_b)
            self.m = self.m - (self.lr * loss_slope_m)
        print(self.m,self.b)
    def predict(self,X):
        return self.m*X+self.b
model=SGD_reg()
model.fit(X_train,Y_train)
y_pred_sgd_reg=model.predict(X_test)
plt.scatter(X,Y,color='blue',label='Data points')
plt.plot(X_test,y_pred_lr,color='red',label='LR',linewidth=5)
plt.plot(X_test,y_pred_sgd_reg,color='silver',label='SGD from scratch')
plt.legend()
```
