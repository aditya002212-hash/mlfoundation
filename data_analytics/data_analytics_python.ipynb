```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np 
```

```python
titanic=pd.read_csv('C:\\Users\\adity\\OneDrive\\Documents\\GitHub\\mlfoundation\\data_analytics\\tested.csv')
```

```python
#4 data describe to get numerical analysis
print(titanic.describe())
```

```python
# 5 null values 
print(titanic.isnull().sum())
```

```python
print(titanic.duplicated().sum())
```

```python
# cabin has 327 null values droping it 

# replacing null of age with median 
titanic['Age']=titanic['Age'].fillna(titanic['Age'].median())
print(titanic)
```

```python
print(titanic.info())
```

```python
# univarite EDA

# categorical data 
# 1. count of survived and not survived
print(titanic['Survived'].value_counts())
plt.bar(np.array([0,1]),titanic['Survived'].value_counts())
plt.xlabel('Survived')
plt.ylabel('Count')
plt.title('Distribution of Survived')
plt.xticks(np.array([0,1]),['Not Survived','Survived'])
plt.show()
```

```python
# 2 . passenger class distribution
print(titanic['Pclass'].value_counts())
plt.pie(titanic['Pclass'].value_counts(),labels=titanic['Pclass'].unique(),autopct='%1.1f%%')
plt.title('Distribution of Passenger Class')
plt.show()
```

```python
# 3, gender wise distribution 
print(titanic['Sex'].value_counts())
plt.pie(titanic['Sex'].value_counts(),labels=titanic['Sex'].unique(),autopct='%1.1f%%')
plt.title('Distribution of Gender')
plt.show()
```

```python
# embarked distribution
print(titanic['Embarked'].value_counts())
plt.bar(titanic['Embarked'].value_counts().index,titanic['Embarked'].value_counts())
plt.xlabel('Port of Embarkation')
plt.ylabel('Count')
plt.title('Distribution of Port of Embarkation')
plt.show()
```
